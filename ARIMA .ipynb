```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
from math import sqrt

# Load dataset
data = pd.read_csv('Cleaned_MCS_TEAM_DATAX.csv')

# Convert 'DATETIME' to datetime object
data['DATETIME'] = pd.to_datetime(data['DATETIME'])

# Sort dataset by date
data = data.sort_values(by='DATETIME')
print(data.columns)


# Encode categorical variables
le_condition = LabelEncoder()
le_flood = LabelEncoder()
data['ACTUAL CONDITION'] = le_condition.fit_transform(data['ACTUAL CONDITION'])
data['OCCURENCE OF FLOOD'] = le_flood.fit_transform(data['OCCURENCE OF FLOOD'])

# Set DATETIME as index
data.set_index('DATETIME', inplace=True)

# Resample to hourly data (assuming original data is at an interval of every 8 hours)
data = data.resample('h').mean().interpolate()  # Interpolating to fill gaps

# Features to forecast
features = ['TEMPERATURE (Â°C)', 'RAINFALL (mm)', 'ACTUAL CONDITION', 'OCCURENCE OF FLOOD']

# Function to fit SARIMA model and make forecasts
def fit_sarima(data, feature, order, seasonal_order, steps):
    model = SARIMAX(data[feature], order=order, seasonal_order=seasonal_order)
    results = model.fit(disp=True)
    forecast = results.forecast(steps)
    return forecast

# Parameters for SARIMA model (p, d, q) and seasonal parameters (P, D, Q, S)
p, d, q = 1, 1, 1  # Example parameters for non-seasonal part
P, D, Q, S = 1, 1, 1, 24  # Example parameters for seasonal part (e.g., daily seasonality)

# Forecast the next 7 days (7 * 24 = 168 hours) for each feature
n_future = 168  # 7 days of hourly data
forecasts = {}

for feature in features:
    forecasts[feature] = fit_sarima(data, feature, (p, d, q), (P, D, Q, S), n_future)

# Create a DataFrame for the forecasted data
forecast_dates = pd.date_range(start=data.index.max() + pd.Timedelta(hours=8), periods=n_future, freq='8H')
forecast_df = pd.DataFrame(forecasts, index=forecast_dates)

# Plotting the forecast results
plt.figure(figsize=(14, 10))
for feature in features:
    plt.subplot(2, 2, features.index(feature) + 1)
    plt.plot(data[feature], label='Historical', color='blue')
    plt.plot(forecast_df[feature], label='Forecast', color='orange')
    plt.title(f'Forecast for {feature}')
    plt.xlabel('Date')
    plt.ylabel(feature)
    plt.legend()

plt.tight_layout()
plt.show()

```

```python
# Load actual data for comparison
actual_data = pd.read_csv('Actual.csv')
actual_data['DATETIME'] = pd.to_datetime(actual_data['DATETIME'])

# Plotting the actual vs forecasted data
plt.figure(figsize=(14, 12))
print(forecast_df.index)
# Plot Temperature
plt.subplot(4, 1, 1)
plt.plot(actual_data['DATETIME'], actual_data['TEMPERATURE (Â°C)'], label='Actual Temperature', color='blue')
plt.plot(forecast_df.index, forecast_df['TEMPERATURE (Â°C)'], label='Forecasted Temperature', color='orange')
plt.title('Temperature Forecast vs Actual')
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)')
plt.legend()

# Plot Rainfall
plt.subplot(4, 1, 2)
plt.plot(actual_data['DATETIME'], actual_data['RAINFALL (mm)'], label='Actual Rainfall', color='blue')
plt.plot(forecast_df, forecast_df['RAINFALL (mm)'], label='Forecasted Rainfall', color='orange')
plt.title('Rainfall Forecast vs Actual')
plt.xlabel('Date')
plt.ylabel('Rainfall (mm)')
plt.legend()

# Plot Actual Condition
plt.subplot(4, 1, 3)
plt.plot(actual_data['DATETIME'], actual_data['ACTUAL CONDITION'], label='Actual Condition', color='blue', marker='o', markersize=2)
plt.plot(forecast_df.iloc[:, 0], forecast_df['ACTUAL CONDITION'], label='Forecasted Condition', color='orange', marker='x', markersize=2)
plt.title('Actual Condition vs Forecasted Condition')
plt.xlabel('Date')
plt.ylabel('Actual Condition')
plt.xticks(rotation=45)
plt.legend()

# Plot Occurrence of Flood
plt.subplot(4, 1, 4)
plt.plot(actual_data['DATETIME'], actual_data['OCCURENCE OF FLOOD'], label='Actual Occurrence of Flood', color='blue', marker='o', markersize=2)
plt.plot(forecast_df.iloc[:, 0], forecast_df['OCCURENCE OF FLOOD'], label='Forecasted Occurrence of Flood', color='orange', marker='x', markersize=2)
plt.title('Occurrence of Flood vs Forecasted Occurrence of Flood')
plt.xlabel('Date')
plt.ylabel('Occurrence of Flood')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
