```python
import pandas as pd

# Load the Excel file and read the first sheet
file_path = 'MCS TEAM (DATA).xlsx'  # Replace with your file path
data = pd.read_excel(file_path, sheet_name='Sheet1')

# Display the first few rows to understand the structure
print("Initial Data:")
print(data.head())

# Step 1: Fill missing 'DATE' values using forward fill (assuming sequential data)
data['DATE'] = data['DATE'].fillna(method='ffill')

# Step 2: Combine 'DATE' and 'TIME' into a single 'DATETIME' column for time-series analysis
data['DATETIME'] = pd.to_datetime(data['DATE'].astype(str) + ' ' + data['TIME'].astype(str))

# Step 3: Set 'DATETIME' as the index and drop the original 'DATE' and 'TIME' columns
data.set_index('DATETIME', inplace=True)
data.drop(columns=['DATE', 'TIME'], inplace=True)

# Display cleaned data
print("Cleaned Data:")
print(data.head())

# Step 4: Optional - Check for and handle missing values in other columns
# This step will fill any remaining NaN values with 0 or a method of your choice
data.fillna(0, inplace=True)

# Step 5: Optionally convert the categorical columns to numeric for further processing
from sklearn.preprocessing import LabelEncoder

# Initialize label encoders for categorical columns
condition_encoder = LabelEncoder()
flood_encoder = LabelEncoder()

# Encode 'ACTUAL CONDITION' and 'OCCURENCE OF FLOOD' into numeric form
data['ACTUAL CONDITION'] = condition_encoder.fit_transform(data['ACTUAL CONDITION'])
data['OCCURENCE OF FLOOD'] = flood_encoder.fit_transform(data['OCCURENCE OF FLOOD'])

# Display the final cleaned data
print("Final Cleaned Data:")
print(data.head())

# Step 6: Save the cleaned dataset as a new CSV file
cleaned_file_path = 'Cleaned_MCS_TEAM_DATA.csv'  # Adjust the file path
data.to_csv(cleaned_file_path, index=True)

```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve

# Load the cleaned dataset
data = pd.read_csv('Cleaned_MCS_TEAM_DATA.csv', index_col='DATETIME', parse_dates=True)

# Prepare features (X) and target (y)
X = data.drop(columns=['OCCURENCE OF FLOOD'])
y = data['OCCURENCE OF FLOOD']

# Split the data into training (80%) and testing (20%)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Normalize the feature data (LSTM requires scaled inputs)
scaler = MinMaxScaler(feature_range=(0, 1))
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Save the scaler for future use
scaler_file = 'scaler.pkl'
joblib.dump(scaler, scaler_file)

# Reshape the input data to be 3D [samples, timesteps, features]
X_train_scaled = np.reshape(X_train_scaled, (X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_scaled = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

# Create DataFrames for training and testing data
X_train_df = pd.DataFrame(X_train, columns=X.columns)
y_train_df = pd.DataFrame(y_train, columns=['OCCURENCE OF FLOOD'])
X_test_df = pd.DataFrame(X_test, columns=X.columns)
y_test_df = pd.DataFrame(y_test, columns=['OCCURENCE OF FLOOD'])

# Concatenate features and target for better visualization
train_data = pd.concat([X_train_df, y_train_df], axis=1)
test_data = pd.concat([X_test_df, y_test_df], axis=1)

# Print training data and testing data as tables
print("Training Data:")
print(train_data.head())  # Print first few rows of training data

print("\nTesting Data:")
print(test_data.head())    # Print first few rows of testing data

# Define the LSTM model
model = Sequential()

# Add LSTM layers with Dropout for regularization
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train_scaled.shape[1], X_train_scaled.shape[2])))
model.add(Dropout(0.2))

model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))

# Add a Dense output layer
model.add(Dense(units=1, activation='sigmoid'))  # Use 'sigmoid' for binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train_scaled, y_train, epochs=50, batch_size=32, validation_data=(X_test_scaled, y_test), verbose=2)

# Predict flood occurrence on the test set
y_pred = model.predict(X_test_scaled)
y_pred = (y_pred > 0.5).astype(int)

# Calculate accuracy, precision, recall, F1 score, and ROC-AUC
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# For ROC-AUC, we need predicted probabilities
y_prob = model.predict(X_test_scaled)
roc_auc = roc_auc_score(y_test, y_prob)

# Print the calculated metrics
print(f"Test Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"ROC-AUC: {roc_auc:.4f}")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['No Flood', 'Flood'], yticklabels=['No Flood', 'Flood'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Plot ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_prob)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Random classifier line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()

# Plot the training and validation loss
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss During Training and Validation')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot the training and validation accuracy
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy During Training and Validation')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

import pandas as pd

# Create a dictionary of the metrics and their corresponding values
metrics_dict = {
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'Score': [accuracy, precision, recall, f1]
}

# Convert the dictionary into a pandas DataFrame
metrics_df = pd.DataFrame(metrics_dict)

# Print the DataFrame as a table
print(metrics_df)

# Optional: Display the DataFrame in a more readable format with a table style (if in a Jupyter environment)
from tabulate import tabulate
print(tabulate(metrics_df, headers='keys', tablefmt='pretty'))

# Plot the model metrics (Accuracy, Precision, Recall, F1-Score)
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
scores = [accuracy, precision, recall, f1]

plt.figure(figsize=(10, 5))

# Create the bar plot with distinct colors
bars = plt.bar(metrics, scores, color=['blue', 'green', 'orange', 'purple'])

# Set the y-axis limit to be between 0 and 1 for score range
plt.ylim(0, 1.1)

# Add title and labels
plt.title('Model Performance Metrics')
plt.ylabel('Score')

# Add value labels on top of each bar with smaller text and better positioning
for bar in bars:
    yval = bar.get_height()  # Get the height of the bar
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f"{yval:.2f}", 
             ha='center', fontweight='bold', fontsize=10)  # Reduced fontsize

# Show the plot
plt.show()




# Save the trained model for future use
model.save('trained_lstm_model.h5')
print("Model saved as 'trained_lstm_model.h5'")



```
